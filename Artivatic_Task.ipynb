```python
# Dropping Columns based on missing data being more than 90% 
a=train_small.isnull().sum()/len(train_small)*100
var = train_small.columns
variable = [ ]
for i in range(0,len(train_small.columns)):
    if a[i]<50:   #setting the threshold as 90% (MissingValue)
        variable.append(var[i])
 # Dropping Columns based on missing data being more than 60% 
drop_col1=list(set(var) - set(variable))
train1=train_small.drop(columns=drop_col1)
```

```python
from sklearn.preprocessing import LabelEncoder

def Encoder(train1):
    columnsToEncode = list(train1.select_dtypes(include=['category','object']))
    le = LabelEncoder()
    for feature in columnsToEncode:
        try:
            train1[feature] = le.fit_transform(train1[feature])
        except:
            print('Error encoding '+feature)
    return train1
train1 = Encoder(train1)
```
